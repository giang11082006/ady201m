2. DATA CLEANING

```python
df = pd.read_csv(('california_final_raw.csv'),sep=',',on_bad_lines='skip',  engine="python")
df.head()
```

```python
# Số lượng data
len(df)
```

```python
nan_value = float("NaN")
df.replace("0", nan_value, inplace=True)
# Quick stats of percentage of null values in columns
(df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
```

```python
#Đổi tất cả giá trị -1 thành NaN
df.replace(-1, np.nan, inplace=True)
```

```python
# Drop cột null > 50%
thresh = len(df) * 0.5
df.dropna(thresh = thresh, axis = 1, inplace = True)
```

```python
#Xóa các cột có giá trị giống hệt nhau
df = df.loc[:, ~df.T.duplicated()]
```

```python
# Drop duplicate columns(same values)
drop_cols = []
for i in range(len(df.columns)):
    for j in range(i+1, len(df.columns)):
        if df[df.columns[i]].equals(df[df.columns[j]]):
            drop_cols.append(df.columns[j])

df.drop(drop_cols, axis=1, inplace=True)
```

```python
#Null percentage after dropping
(df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
```
