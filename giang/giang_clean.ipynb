```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("california_master_dataset_1111.csv")
df.head(5)
```

```python
def add_one(x):
    """
    Trả về x + 1.
    Hỗ trợ số (int/float/np.number), numpy array, pandas Series hoặc DataFrame.
    """
    try:
        return x + 1
    except Exception as e:
        raise TypeError(f"Không thể cộng 1 cho đối tượng loại {type(x)}: {e}")
```

```python
df.shape
```

```python
usefull_cols = [
    'zpid',
    'hdpData.homeInfo.price', 
    'statusText', 
    'hdpData.homeInfo.homeType',
    'hdpData.homeInfo.city',
    'hdpData.homeInfo.state', 
    'hdpData.homeInfo.latitude', 
    'hdpData.homeInfo.longitude',
    'baths', 
    'beds', 
    'area',
    'hdpData.homeInfo.lotAreaValue', 
    'hdpData.homeInfo.lotAreaUnit',
    'timeOnZillow'
]

df = df[usefull_cols]
```

```python
df.rename(columns={
    'hdpData.homeInfo.price': 'price',
    'statusText': 'status',
    'hdpData.homeInfo.homeType': 'type',
    'hdpData.homeInfo.city': 'city',
    'hdpData.homeInfo.state': 'state',
    'hdpData.homeInfo.latitude': 'latitude',
    'hdpData.homeInfo.longitude': 'longitude',
    'baths': 'bathrooms',
    'beds': 'bedrooms',
    'area': 'living_area',
    'hdpData.homeInfo.lotAreaValue': 'lot_area',
    'hdpData.homeInfo.lotAreaUnit': 'lot_area_unit',
    'timeOnZillow': 'days_on_zillow'
}, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# CÓ 58 NHÀ DIỆN TÍCH BẰNG 0 , ĐỔI RA NAN RỒI FILL MEDIAN
print((df['lot_area'] == 0).sum())
```

```python
mask = df['lot_area'] < 20

df.loc[mask, 'lot_area'] *= 43560
```

```python
df = df.dropna(subset=["zpid"])
```

```python
df.head(5)
```

```python
# Xử lý null bằng cách fill  vào bằng median
df["bedrooms"] = df["bedrooms"].fillna(
    df["bedrooms"].median()
)

df["bathrooms"] = df["bathrooms"].fillna(
    df["bathrooms"].median()
)
```

```python
print(df['lot_area'].min())
print(df['lot_area'].max())
```

```python
print(df.isnull().sum())
```

```python
print((df["living_area"] == 0).sum())

print((df["lot_area"] == 0).sum())
```

```python
df["living_area"] = df["living_area"].replace(0, np.nan)
df["lot_area"] = df["lot_area"].replace(0, np.nan)
```

```python
print(df.isnull().sum())
```

```python
df["living_area"] = df["living_area"].fillna(
    df["living_area"].median()
)

df["lot_area"] = df["lot_area"].fillna(
    df["lot_area"].median()
)
```

```python
print(df.isnull().sum())
```

```python
df.drop(columns=["lot_area_unit"], inplace=True)
```
