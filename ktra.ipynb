```python
import pandas as pd
import os

# 1. Định vị thư mục làm việc
workspace = "./"
# 2. Khai báo danh sách 3 file cần ghép
file_paths = [
    os.path.join(workspace, "zillow_mass_dataset_full.csv"),
    os.path.join(workspace, "Bản sao zillow_national_dataset_CA_1.csv"),
    os.path.join(workspace, "zillow_national_data_CA_2.csv")
]

output_path = os.path.join(workspace, "california_master_dataset_1111.csv")

# 3. Đọc và gom tất cả vào một list các DataFrame
print("Đang đọc và gộp dữ liệu...")
df_list = []

for file in file_paths:
    if os.path.exists(file):
        df = pd.read_csv(file)
        df_list.append(df)
        print(f"  -> Đã nạp: {os.path.basename(file)} ({len(df)} dòng)")
    else:
        print(f"  [!] Không tìm thấy file: {file}")

# 4. Gộp toàn bộ lại thành 1 DataFrame duy nhất
# ignore_index=True để đánh số thứ tự (index) lại từ 0 cho mượt
df_master = pd.concat(df_list, ignore_index=True)

# 5. Xóa các căn nhà bị trùng lặp (dựa trên zpid)
if 'zpid' in df_master.columns:
    truoc = len(df_master)
    df_master = df_master.drop_duplicates(subset=['zpid'], keep='first')
    sau = len(df_master)
    print(f"\n[*] Đã dọn dẹp {truoc - sau} căn nhà bị trùng lặp giữa 3 file.")

# 6. Xuất ra file Master
df_master.to_csv(output_path, index=False, encoding="utf-8-sig")

print("="*60)
print(f"[THÀNH CÔNG] File tổng hợp đã ra lò tại: {output_path}")
print(f"Tổng số dữ liệu sạch: {len(df_master)} dòng.")
```
